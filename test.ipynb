```python
import pandas as pd

df_login = pd.read_csv("login.csv")

df_login.head()

data_login = {'usernames' : {}}
colonne_list = df_login.columns

for i in range(len(df_login)) :
    data_user = {}
    for j in range(len(colonne_list)) :
        colonne =  colonne_list[j]
        data_user[colonne] = df_login.iloc[i][j]  
    data_login['usernames'][df_login.iloc[i][0]] = data_user

data_login
```
